```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}

def get_owner():
    number = input('Введите номер документа:   ')
    for dic in documents:
        if "number" in dic.keys() and dic["number"] == number:
            print('Владелец документа:       ', dic['name'])   
            break
    else:
          print('Документ не найден в базе')
            
def get_shelf():
    number = input('Введите номер документа:   ')
    for shelf_id, list_of_nums in directories.items():
        if number in list_of_nums:
            print ('Документ хранится на полке', shelf_id)
            break
    else:
            print ('Документ не найден в базе' )
            

def get_shelf_num(doc_num):
    for shelf_id, list_of_nums in directories.items():
        if doc_num in list_of_nums:
            return shelf_id

def l():
    for info in documents:
        shelf = get_shelf_num(info['number'])
        print(f"№: {info['number']}, тип: {info['type']}, владелец: {info['name']}, полка хранения: {shelf}")

def add_new_shelf():
    shelf_num = input("Введите номер полки: ")
    if shelf_num in directories.keys():
        print(f"Такая полка уже существует. Текущий перечень полок: {', '.join(directories.keys())}")
    else:
        directories[shelf_num] = []
        print(f"Полка добавлена. Текущий перечень полок: {', '.join(directories.keys())}")

def del_shelf_if_empty():
    shelf_num = input("Введите номер полки: ")
    
    if shelf_num in directories and len(directories[shelf_num]) == 0:
            del(directories[shelf_num])
            print(f"Полка удалена. Текущий перечень полок: {', '.join(directories.keys())}") 
    elif shelf_num in directories:
            print(f"На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: {', '.join(directories.keys())}")
    else:
            print (f"Такой полки не существует.Текущий перечень полок: {', '.join(directories.keys())}")
            
           
     
     
    
comm = input("Введите команду: ")
while comm != 'exit':
    if comm == 'p':
        get_owner()
    elif comm == 's':
        get_shelf()
    elif comm == 'l':
        l()
    elif comm == 'as':
        add_new_shelf()
    elif comm == 'ds':
        del_shelf_if_empty()
    else:
        print('Неизвестная команда. Для выхода введите команду exit')
    comm = input("Введите команду: ")
print('Работа завершена')
```
