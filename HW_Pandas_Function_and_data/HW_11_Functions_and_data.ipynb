# Задание 1
Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
- оценка 2 и меньше - низкий рейтинг
- оценка 4 и меньше - средний рейтинг
- оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class

```python
import pandas as pd
import pandas as np
ratings111 = pd.read_csv('ratings111.csv')
print (ratings111.head(1))
movies = pd.read_csv('movies.csv')
print (movies.head(1))
```

```python
# создаю общую таблицу Total со слиянием df при помощи how=inner указав правый df movies и ключ соединения movieId
Total = ratings111.merge(movies, on=["movieId"], how='inner')
```

```python
# использую метод pivot_table указав индекс,значение и параметр aggfunc задающий тип агрегации
df = Total.pivot_table(index=['title','movieId'], values = 'rating', aggfunc='mean')
df.head()
```

```python
# сбрасываю индекс и использую новый, добавляя старый как столбец 
df = df.reset_index()
```

```python
# функция классификации по значению рейтинга
def classification(value_rating):
    if value_rating  <= 2:
        return 'низкий рейтинг'
    elif value_rating  <= 4 and value_rating > 2:
        return 'средний рейтинг'
    elif value_rating  > 4.5:
        return 'высокий рейтинг'
```

```python
# использую метод apply создавая новый столбец class и classs(добавил одну s,что бы дальше поработать)
df['classs'] = df['rating'].apply(classification)
df.head(10)
```

```python
# решил вывести информацию по каждому рейтингу
df_Low_rating = df[df.classs == 'низкий рейтинг']
df_Low_rating
```

```python
df_Mid_rating = df[df.classs == 'средний рейтинг']
df_Mid_rating
```

```python
df_High_rating = df[df.classs == 'высокий рейтинг']
df_High_rating
```

Используем файл keywords.csv.

Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определенному региону. Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’ пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим ‘undefined’.

Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:

geo_data = {

'Центр': ['москва', 'тула', 'ярославль'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}

Результат классификации запишите в отдельный столбец region.

```python
import pandas as pd
import pandas as np

df = pd.read_csv('keywords.csv')
print(df.head())  
print(df.tail())
```

```python
#Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:
geo_data = {
    
    'Центр': ['москва', 'тула', 'ярославль'],
    
    'Северо-Запад': ['петербург', 'псков', 'мурманск'],
    
    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск'] 
}
```

```python
# создаю функцию с вариантами распределения по регионам и  вариантом по умолчанию дающим undefined
# почему-то выдает ошибку если вместо or использовать | -причину не нашел
def geo_classifier(row):
    if 'москва' in row['keyword'] or 'тула' in row['keyword'] or 'ярославль' in row['keyword']:
        return 'Центр'
    elif 'петербург' in row['keyword'] or 'псков' in row['keyword'] or 'псков' in row['keyword']:
        return 'Северо-Запад'
    elif 'владивосток' in row['keyword'] or 'сахалин' in row['keyword'] or 'хабаровск' in row['keyword']:
        return 'Дальний Восток'
    else:
        return 'undefined'
    
```

```python
# использую метод apply ко всему df и прохожу по нему всему сверху вниз передавая всю строку в виде df из одной строки
df['region'] = df.apply(geo_classifier, axis=1)
print(df.head())  
print(df.tail())
```

```python
# решил вывести информацию по каждому региону
df_Centr = df[df.region == 'Центр']
df_Centr
```

```python
df_North_west = df[df.region == 'Северо-Запад']
df_North_west
```

```python
df_Far_East = df[df.region == 'Дальний Восток']
df_Far_East
```
