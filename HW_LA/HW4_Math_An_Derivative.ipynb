## –î–æ–º–∞—à–Ω–µ–µ –∑–∞–¥–∞–Ω–∏–µ "–ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–∏–π –∞–Ω–∞–ª–∏–∑. –ü—Ä–æ–∏–∑–≤–æ–¥–Ω–∞—è".
#### –£—Ä–æ–≤–µ–Ω—å 0:  
–ü–æ—Å—á–∏—Ç–∞–π—Ç–µ –∑–Ω–∞—á–µ–Ω–∏–µ –ø–µ—Ä–≤–æ–π –∏ –≤—Ç–æ—Ä–æ–π –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–∏:

$f(x) = x^5 + 4sin(2x) + cos(3x+3)$

–í —Ç–æ—á–∫–µ x = 1.

#### –£—Ä–æ–≤–µ–Ω—å 1:  
–ü–æ—Å—Ç—Ä–æ–π—Ç–µ –≥—Ä–∞—Ñ–∏–∫ –ø–µ—Ä–≤–æ–π –∏ –≤—Ç–æ—Ä–æ–π –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–∏:

$f(x) = sin^5(2x+1)$

–ù–∞ –æ—Ç—Ä–µ–∑–∫–µ [-5; 5].

## –£—Ä–æ–≤–µ–Ω—å 0:
–ü–æ—Å—á–∏—Ç–∞–π—Ç–µ –∑–Ω–∞—á–µ–Ω–∏–µ –ø–µ—Ä–≤–æ–π –∏ –≤—Ç–æ—Ä–æ–π –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–∏:

ùëì(ùë•)=ùë•5+4ùë†ùëñùëõ(2ùë•)+ùëêùëúùë†(3ùë•+3)   
–í —Ç–æ—á–∫–µ x = 1.

```python
import numpy as np
from sympy import *
from scipy.misc import derivative
```

```python
x, y = symbols('x y') 
gfg_ex = x**5 + 4*sin(2*x) + cos(3*x+ 3)
```

```python
# –Ω–∞—Ö–æ–∂—É –ø–µ—Ä–≤—É—é –ø—Ä–æ–∏–∑–≤–æ–¥–Ω—É—é(–¥–ª—è –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏–∏)
dif = diff(gfg_ex, x) 
dif
```

```python
# –Ω–∞—Ö–æ–∂—É –ø–µ—Ä–≤—É—é –ø—Ä–æ–∏–∑–≤–æ–¥–Ω—É—é –≤ —Ç–æ—á–∫–µ x = 1
def f(x):
    return x**5 + 4*sin(2*x) + cos(3*x + 3)
derivative(f, 1.0, dx = 1e-6, n = 1)
```

```python
# –Ω–∞—Ö–æ–∂—É –≤—Ç–æ—Ä—É—é –ø—Ä–æ–∏–∑–≤–æ–¥–Ω—É—é(–¥–ª—è –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏–∏)
dif2 = diff(dif, x) 
dif2
```

```python
# –Ω–∞—Ö–æ–∂—É –≤—Ç–æ—Ä—É—é –ø—Ä–æ–∏–∑–≤–æ–¥–Ω—É—é –≤ —Ç–æ—á–∫–µ x = 1
def f(x):
    return x**5 + 4*sin(2*x) + cos(3*x + 3)
derivative(f, 1.0, dx = 1e-6, n = 2)
```

## –£—Ä–æ–≤–µ–Ω—å 1:
–ü–æ—Å—Ç—Ä–æ–π—Ç–µ –≥—Ä–∞—Ñ–∏–∫ –ø–µ—Ä–≤–æ–π –∏ –≤—Ç–æ—Ä–æ–π –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–∏:  
ùëì(ùë•)=ùë†ùëñùëõ5(2ùë•+1)   
–ù–∞ –æ—Ç—Ä–µ–∑–∫–µ [-5; 5]

```python
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.sin(2*x + 1)**5
```

```python
x = np.linspace(-5, 5)
fx = f(x)
f1x = [derivative(f, xx, dx = 1e-6, n = 1) for xx in x]
```

```python
# –≥—Ä–∞—Ñ–∏–∫ —Ñ—É–Ω–∫—Ü–∏–∏ –∏ 1-–æ–π –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–æ–π
plt.plot(x, fx, label='function')
plt.plot(x, f1x, label='derivative')
plt.legend()
plt.show()
```

```python
x = np.linspace(-5, 5)
fx = f(x)
f1x = [derivative(f, xx, dx = 1e-6, n = 2) for xx in x]
```

```python
# –≥—Ä–∞—Ñ–∏–∫ —Ñ—É–Ω–∫—Ü–∏–∏ –∏ 2-–æ–π –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–æ–π
plt.plot(x, fx, label='function')
plt.plot(x, f1x, label='derivative')
plt.legend()
plt.show()
```
