```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as sts
from scipy.stats import norm
import random
```

```python
for vals in [(2.0, 0.5), (1.5, 0.5), (2.5, 0.5), (2.0, 0.45), (2.0, 0.55)]:
    norm_rv = sts.norm(loc = vals[0], scale = vals[1])
    x = np.linspace(0, 4, 100)
    y = norm_rv.pdf(x)
    plt.plot(x, y, label = f"µ = {vals[0]}, σ = {vals[1]}")
plt.title("Нормальное распределение")
plt.ylabel("f(x)")
plt.xlabel("x")
plt.legend(loc = 'best', frameon = False)
plt.grid()
```

```python
for vals in [2, 3, 4, 5, 6]:
    expon_rv = sts.expon(vals)
    x1 = np.linspace(0, 12, 100)
    y1 = expon_rv.pdf(x1)
    plt.plot(x1, y1, label = f"lam = {vals}")
plt.title("Экспоненциальное распределение")
plt.ylabel("f(x1)")
plt.xlabel("x1")
plt.legend(loc = 'best', frameon = False)
plt.grid()
```

```python
for vals in [0.1, 0.5 , 1.0 , 1.5 , 2.0 ]:
    t_rv = sts.t(vals)
    x2 = np.linspace(-5, 5, 1000)
    y2 = t_rv.pdf(x2)
    plt.plot(x2, y2, label = f"n = {vals}")
plt.title("Распределение Стьюдента")
plt.ylabel("f(x2)")
plt.xlabel("x2")
plt.legend(loc = 'best', frameon = False)
plt.grid()
```
