### 1.3. Описание
Данные из Информационной службы развивающихся рынков (EMIS, [Web Link]), которая представляет собой базу данных, содержащую информацию о развивающихся рынках по всему миру. Компании-банкроты были проанализированы в период с 2000 по 2012 год, а компании, которые еще работают - с 2007 по 2013 годы.

На основе собранных данных выделено пять случаев классификации в зависимости от периода прогнозирования:

* 1-й год(1stYear): данные содержат финансовые ставки с 1-го года прогнозного периода и соответствующую метку класса, которая указывает на статус банкротства через 5 лет. Данные содержат 7027 экземпляров (финансовая отчетность), 271 - обанкротившиеся компании, 6756 фирм не обанкротились в прогнозный период.

* 2-й год(2ndYear): данные содержат финансовые показатели со 2-го года периода прогнозирования и соответствующую метку класса, которая указывает на статус банкротства через 4 года. Данные содержат 10173 экземпляра (финансовая отчетность), 400 - обанкротившиеся компании, 9773 фирмы, не являвшиеся банкротами в прогнозный период.

* 3-й год(3rdYear): данные содержат финансовые показатели с 3-го года прогнозного периода и соответствующую метку класса, которая указывает на статус банкротства через 3 года. Данные содержат 10503 экземпляра (финансовая отчетность), 495 - обанкротившиеся компании, 1 0008 фирм, не являвшихся банкротами в прогнозный период.

* 4-й год(4thYear): данные содержат финансовые показатели с 4-го года прогнозного периода и соответствующую метку класса, которая указывает на статус банкротства через 2 года. * 5-й год(5thYear): данные содержат финансовые ставки с 5-го года прогнозного периода и соответствующую метку класса, которая указывает на статус банкротства через 1 год. * Accuracy: отношение правильно спрогнозированных наблюдений к общему количеству наблюдений(доля от общего числа верных прогнозов)   
  ** Accuracy = TP + TN / TP + FP + FN + TN
* Precision: отношение правильно предсказанных положительных наблюдений к общему количеству предсказанных положительных наблюдений(доля действительно верных положительных прогнозов).  
  ** Precision = TP / TP + FP  
* Recall: отношение правильно предсказанных положительных наблюдений ко всем наблюдениям (доля положительных наблюдаемых значений, правильно предсказанных).  
  ** Recall = TP / TP + FN
* F1 score: представляет собой средневзвешенное значение Recall and Precision
  ** F1 Score = 2*(Recall * Precision) / (Recall + Precision)

##### Для нашей задачи:

* Accuracy:  в целом, как часто классификатор верен.
* Precision: доля действительно верных прогнозов банкротства.
* Recall:    доля обанкротившихся компаний, которые фактически определяет классификатор.
* F1: средневзвешенное значение доли верных прогнозов банкротсва и фактически обанкротившихся компаний
